# Logistic ODE to Logistic Map

$$ p_{n+1} = p_n + kp_n(1-\frac{p_n}{C}) $$

Multiply everything by C

$$ C*p_{n+1} = C*p_n + C*kp_n(1-\frac{p_n}{C}) $$

Let $x_n = \frac{p_n}{C}$ and plug it in to give us:

$$ x_{n+1} = x_n + kx_n(1-x_n) $$

Factor out $x_n$

$$ x_{n+1} = x_n(1 + k - kx_n)$$

$$x_{n+1} = -kx_n^2 + (1+k)x_n $$

Multiply everything by $\frac{k}{k+1}$

$$\frac{k}{k+1}x_{n+1} = \frac{k}{k+1}(-kx_n^2 + (1+k)x_n) $$

$$\frac{k}{k+1}x_{n+1} = -\frac{k^2}{k+1}x_n^2 + kx_n $$

Now if we let $l_n = \frac{k+1}{k}x_n$ then we can plug that in and get

$$l_{n+1} = -(k+1)l_n^2 + (k+1)l_n $$

$$l_{n+1} = (k+1)l_n(-l_n + 1) $$

$$l_{n+1} = (k+1)l_n(1 - l_n) $$

Let r = k+1 and we end up with the logistic map!

$$l_{n+1} = rl_n(1-l_n) $$
